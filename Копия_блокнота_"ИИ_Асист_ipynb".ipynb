```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
import os

def preprocess_files(directory, encryption_key):
    """
    –§—É–Ω–∫—Ü–∏—è –æ–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ—Ç –≤—Å–µ —Ñ–∞–π–ª—ã –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –¥–∏—Ä–µ–∫—Ç–æ—Ä–∏–∏ –∏ –≤–æ –≤—Å–µ—Ö –≤–ª–æ–∂–µ–Ω–Ω—ã—Ö –ø–∞–ø–∫–∞—Ö.
    –î–ª—è PDF —Ñ–∞–π–ª–æ–≤ –∏–∑–≤–ª–µ–∫–∞–µ—Ç —Ç–µ–∫—Å—Ç, –¥–ª—è PPTX —Ñ–∞–π–ª–æ–≤ - –∏–∑–≤–ª–µ–∫–∞–µ—Ç —Ç–µ–∫—Å—Ç –∏–∑ —Å–ª–∞–π–¥–æ–≤.
    –í–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ª–æ–≤–∞—Ä—å, –≥–¥–µ –∫–ª—é—á–∏ - –∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω—ã–µ –∏–º–µ–Ω–∞ —Ñ–∞–π–ª–æ–≤, –∞ –∑–Ω–∞—á–µ–Ω–∏—è - –∑–∞—à–∏—Ñ—Ä–æ–≤–∞–Ω–Ω—ã–π –∏–∑–≤–ª–µ—á–µ–Ω–Ω—ã–π —Ç–µ–∫—Å—Ç.
    """
    file_text = {}
    for root, dirs, files in os.walk(directory):
        for filename in files:
            if filename.endswith(".pdf"):
                pdf_path = os.path.join(root, filename)
                with open(pdf_path, 'rb') as pdf_file:
                    with PDF(pdf_file) as pdf:
                        text = "\n".join(page.extract_text() for page in pdf.pages)
                        encrypted_text = encrypt_text(text, encryption_key)
                        encrypted_filename = encrypt_filename(pdf_path, encryption_key)
                        file_text[encrypted_filename] = encrypted_text
            elif filename.endswith(".pptx"):
                pptx_path = os.path.join(root, filename)
                presentation = pptx.Presentation(pptx_path)
                text_slides = []
                for slide in presentation.slides:
                    text = "\n".join(shape.text for shape in slide.shapes if hasattr(shape, "text"))
                    text_slides.append(text)
                text = "\n\n".join(text_slides)
                encrypted_text = encrypt_text(text, encryption_key)
                encrypted_filename = encrypt_filename(pptx_path, encryption_key)
                file_text[encrypted_filename] = encrypted_text
    return file_text

def encrypt_text(text, key):
    cipher = AES.new(key, AES.MODE_GCM)
    ciphertext, tag = cipher.encrypt_and_digest(text.encode())
    return ciphertext, tag, cipher.nonce

def encrypt_filename(filename, key):
    cipher = AES.new(key, AES.MODE_GCM)
    ciphertext, tag = cipher.encrypt_and_digest(filename.encode())
    return f"{ciphertext.hex()}_{tag.hex()}_{cipher.nonce.hex()}"

# –ü—Ä–∏–º–µ—Ä –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è
encryption_key = get_random_bytes(16)
db_directory = "/content/Db"
file_texts = preprocess_files(db_directory, encryption_key)
```

```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend

def encrypt_data(data, key):
    """
    –®–∏—Ñ—Ä—É–µ—Ç –¥–∞–Ω–Ω—ã–µ, –∏—Å–ø–æ–ª—å–∑—É—è —É–∫–∞–∑–∞–Ω–Ω—ã–π –∫–ª—é—á.
    """
    iv = get_random_bytes(16)
    cipher = Cipher(algorithms.AES(key), modes.GCM(iv), backend=default_backend())
    encryptor = cipher.encryptor()
    ciphertext = encryptor.update(data) + encryptor.finalize()
    tag = encryptor.tag
    return (ciphertext, tag, iv)

def decrypt_data(encrypted_data, key):
    """
    –†–∞—Å—à–∏—Ñ—Ä–æ–≤—ã–≤–∞–µ—Ç –¥–∞–Ω–Ω—ã–µ, –∏—Å–ø–æ–ª—å–∑—É—è —É–∫–∞–∑–∞–Ω–Ω—ã–π –∫–ª—é—á.
    """
    ciphertext, tag, iv = encrypted_data
    cipher = Cipher(algorithms.AES(key), modes.GCM(iv), backend=default_backend())
    decryptor = cipher.decryptor()
    decryptor.authenticate_additional_data(b"")
    return decryptor.update(ciphertext) + decryptor.finalize()
```

```python
import os
from datetime import datetime

def can_access_file(filename, file_size, file_mtime, key):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç, –º–æ–∂–µ—Ç –ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∫–ª—é—á–æ–º –ø–æ–ª—É—á–∏—Ç—å –¥–æ—Å—Ç—É–ø –∫ —Ñ–∞–π–ª—É.
    """
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–æ –∏–º–µ–Ω–∏ —Ñ–∞–π–ª–∞
    if filename.startswith(key.decode()):
        return True

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–æ —Ä–∞–∑–º–µ—Ä—É —Ñ–∞–π–ª–∞
    max_file_size = int(key[0:8])
    if file_size <= max_file_size:
        return True

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–æ –¥–∞—Ç–µ –º–æ–¥–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Ñ–∞–π–ª–∞
    max_mtime = int(key[8:16])
    file_mtime_dt = datetime.fromtimestamp(file_mtime)
    max_mtime_dt = datetime.fromtimestamp(max_mtime)
    if file_mtime_dt <= max_mtime_dt:
        return True

    return False
```

```python
def preprocess_files(directory, encryption_key, file_types=None):
    file_texts = {}
    for filename in os.listdir(directory):
        if file_types is None or any(filename.endswith(ft) for ft in file_types):
            file_path = os.path.join(directory, filename)
            file_size = os.path.getsize(file_path)
            file_mtime = os.path.getmtime(file_path)
            with open(file_path, 'rb') as file:
                file_data = file.read()
            encrypted_data = encrypt_data(file_data, encryption_key)
            file_texts[(filename, file_size, file_mtime)] = encrypted_data
    return file_texts

def manage_access(file_texts, access_permissions):
    accessible_files = {}
    for (filename, file_size, file_mtime), encrypted_data in file_texts.items():
        for user, key in access_permissions.items():
            if can_access_file(filename, file_size, file_mtime, key):
                decrypted_data = decrypt_data(encrypted_data, key)
                accessible_files[(filename, file_size, file_mtime)] = decrypted_data
                break
    return accessible_files
```

```python
def can_access_file(filename, file_size, file_mtime, key, user, action):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç, –º–æ–∂–µ—Ç –ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º –∫–ª—é—á–æ–º –≤—ã–ø–æ–ª–Ω–∏—Ç—å –æ–ø—Ä–µ–¥–µ–ª–µ–Ω–Ω–æ–µ –¥–µ–π—Å—Ç–≤–∏–µ —Å —Ñ–∞–π–ª–æ–º.
    """
    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–æ –∏–º–µ–Ω–∏ —Ñ–∞–π–ª–∞
    if filename.startswith(key.decode()):
        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∞–≤ –¥–æ—Å—Ç—É–ø–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ —Å—Ç—Ä–æ–∫
        if user in user_permissions and action in user_permissions[user]:
            return True

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–æ —Ä–∞–∑–º–µ—Ä—É —Ñ–∞–π–ª–∞
    max_file_size = int(key[0:8])
    if file_size <= max_file_size:
        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∞–≤ –¥–æ—Å—Ç—É–ø–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ —Å—Ç—Ä–æ–∫
        if user in user_permissions and action in user_permissions[user]:
            return True

    # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø–æ –¥–∞—Ç–µ –º–æ–¥–∏—Ñ–∏–∫–∞—Ü–∏–∏ —Ñ–∞–π–ª–∞
    max_mtime = int(key[8:16])
    file_mtime_dt = datetime.fromtimestamp(file_mtime)
    max_mtime_dt = datetime.fromtimestamp(max_mtime)
    if file_mtime_dt <= max_mtime_dt:
        # –ü—Ä–æ–≤–µ—Ä–∫–∞ –ø—Ä–∞–≤ –¥–æ—Å—Ç—É–ø–∞ –Ω–∞ —É—Ä–æ–≤–Ω–µ —Å—Ç—Ä–æ–∫
        if user in user_permissions and action in user_permissions[user]:
            return True

    return False
```

```python
def manage_access(file_texts, access_permissions, user, action):
    """
    –ü—Ä–µ–¥–æ—Å—Ç–∞–≤–ª—è–µ—Ç –¥–æ—Å—Ç—É–ø –∫ —Ñ–∞–π–ª–∞–º –≤ —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–∏–∏ —Å —É–∫–∞–∑–∞–Ω–Ω—ã–º–∏ —Ä–∞–∑—Ä–µ—à–µ–Ω–∏—è–º–∏ –∏ –¥–µ–π—Å—Ç–≤–∏—è–º–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è.
    """
    accessible_files = {}
    for (filename, file_size, file_mtime), encrypted_data in file_texts.items():
        for key in access_permissions.values():
            if can_access_file(filename, file_size, file_mtime, key, user, action):
                decrypted_data = decrypt_data(encrypted_data, key)
                accessible_files[(filename, file_size, file_mtime)] = decrypted_data
                break
    return accessible_files
```

```python
import re

def extract_keywords(text):
    """
    –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ç–µ–∫—Å—Ç –∏–∑ —Ñ–∞–π–ª–∞ –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Å–ª–æ–≤–∞—Ä—å –∫–ª—é—á–µ–≤—ã—Ö —Å–ª–æ–≤ —Å –∏—Ö —á–∞—Å—Ç–æ—Ç–æ–π.
    """
    # –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º —Ç–µ–∫—Å—Ç –≤ –Ω–∏–∂–Ω–∏–π —Ä–µ–≥–∏—Å—Ç—Ä
    text = text.lower()

    # –£–¥–∞–ª—è–µ–º –∑–Ω–∞–∫–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è
    text = re.sub(r'[^\w\s]', '', text)

    # –°–æ–∑–¥–∞–µ–º —Å–ø–∏—Å–æ–∫ —Å—Ç–æ–ø-—Å–ª–æ–≤ –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ
    stop_words = ['–∞', '–∏', '–≤', '–≤–æ', '–Ω–µ', '—á—Ç–æ', '–æ–Ω', '–Ω–∞', '—è', '—Å', '—Å–æ', '–∫–∞–∫', '–¥–ª—è', '—ç—Ç–æ—Ç', '–±—ã—Ç—å', '–ø—Ä–∏', '–æ—Ç', '–ø–æ', '–∏–∑', '–µ–≥–æ', '–µ—ë', '—É–∂–µ', '–∫–∞–∫–æ–π', '—ç—Ç–æ–≥–æ', '—Å–µ–±—è', '–Ω–∏—Ö', '—Ç–æ–π', '–Ω–∞–¥', '–±—ã–ª–∏', '–±—É–¥–µ—Ç', '–µ–µ', '–º—ã', '—Ç—É—Ç', '–µ—Å—Ç—å', '–Ω–µ–µ', '–≤—Å–µ—Ö', '–±—É–¥—É—Ç', '—Ç–æ–≥–æ']

    # –†–∞–∑–±–∏–≤–∞–µ–º —Ç–µ–∫—Å—Ç –Ω–∞ —Å–ª–æ–≤–∞ –∏ —Ñ–∏–ª—å—Ç—Ä—É–µ–º —Å—Ç–æ–ø-—Å–ª–æ–≤–∞
    words = [word for word in re.findall(r'\w+', text) if word not in stop_words]

    # –§–æ—Ä–º–∏—Ä—É–µ–º —Å–ª–æ–≤–∞—Ä—å –∫–ª—é—á–µ–≤—ã—Ö —Å–ª–æ–≤ —Å –∏—Ö —á–∞—Å—Ç–æ—Ç–æ–π
    keywords = {}
    for word in words:
        if word in keywords:
            keywords[word] += 1
        else:
            keywords[word] = 1

    # –°–æ—Ä—Ç–∏—Ä—É–µ–º —Å–ª–æ–≤–∞—Ä—å –ø–æ —á–∞—Å—Ç–æ—Ç–µ –≤ –ø–æ—Ä—è–¥–∫–µ —É–±—ã–≤–∞–Ω–∏—è
    keywords = dict(sorted(keywords.items(), key=lambda item: item[1], reverse=True))

    return keywords
```

```python
import spacy
from collections import Counter
from spacy.tokens import Span

def extract_document_meaning(text):
    """
    –§—É–Ω–∫—Ü–∏—è –ø—Ä–∏–Ω–∏–º–∞–µ—Ç —Ç–µ–∫—Å—Ç –¥–æ–∫—É–º–µ–Ω—Ç–∞ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç –µ–≥–æ –æ—Å–Ω–æ–≤–Ω–æ–π —Å–º—ã—Å–ª.
    """
    # –ó–∞–≥—Ä—É–∂–∞–µ–º –º–æ–¥–µ–ª—å spaCy –¥–ª—è —Ä—É—Å—Å–∫–æ–≥–æ —è–∑—ã–∫–∞
    nlp = spacy.load("ru_core_news_sm")

    # –û–±—Ä–∞–±–∞—Ç—ã–≤–∞–µ–º —Ç–µ–∫—Å—Ç —Å –ø–æ–º–æ—â—å—é spaCy
    doc = nlp(text)

    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –∫–ª—é—á–µ–≤—ã–µ —Å—É—â–Ω–æ—Å—Ç–∏ –≤ —Ç–µ–∫—Å—Ç–µ
    entities = []
    for ent in doc.ents:
        # –û–ø—Ä–µ–¥–µ–ª—è–µ–º —Ç–∏–ø —Å—É—â–Ω–æ—Å—Ç–∏
        ent_type = ent.label_
        if ent_type == 'ORG':
            # –î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –ø—Ä–æ–≤–µ—Ä—è–µ–º, –¥–µ–π—Å—Ç–≤–∏—Ç–µ–ª—å–Ω–æ –ª–∏ —ç—Ç–æ –æ—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏—è
            if any(token.text.lower() in ['–∫–æ–º–ø–∞–Ω–∏—è', '—Ñ–æ–Ω–¥', '—Ü–µ–Ω—Ç—Ä', '–∏–Ω—Å—Ç–∏—Ç—É—Ç', '—É–Ω–∏–≤–µ—Ä—Å–∏—Ç–µ—Ç', '–∑–∞–≤–æ–¥', '–ø—Ä–µ–¥–ø—Ä–∏—è—Ç–∏–µ'] for token in ent):
                ent_type = '–û—Ä–≥–∞–Ω–∏–∑–∞—Ü–∏—è'
            else:
                ent_type = '–î—Ä—É–≥–æ–µ'
        elif ent_type == 'LOC':
            ent_type = '–ú–µ—Å—Ç–æ–ø–æ–ª–æ–∂–µ–Ω–∏–µ'
        elif ent_type == 'PER':
            ent_type = '–ü–µ—Ä—Å–æ–Ω–∞'
        elif ent_type == 'GPE':
            ent_type = '–ì–µ–æ–≥—Ä–∞—Ñ–∏—á–µ—Å–∫–∏–π –æ–±—ä–µ–∫—Ç'
        elif ent_type == 'PRODUCT':
            ent_type = '–ü—Ä–æ–¥—É–∫—Ç'
        elif ent_type == 'EVENT':
            ent_type = '–°–æ–±—ã—Ç–∏–µ'
        elif ent_type == 'WORK_OF_ART':
            ent_type = '–ü—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ –∏—Å–∫—É—Å—Å—Ç–≤–∞'
        elif ent_type == 'LAW':
            ent_type = '–ó–∞–∫–æ–Ω'
        elif ent_type == 'MONEY':
            ent_type = '–í–∞–ª—é—Ç–∞/–î–µ–Ω–µ–∂–Ω–∞—è —Å—É–º–º–∞'
        elif ent_type == 'DATE':
            ent_type = '–î–∞—Ç–∞'
        elif ent_type == 'TIME':
            ent_type = '–í—Ä–µ–º—è'
        elif ent_type == 'PERCENT':
            ent_type = '–ü—Ä–æ—Ü–µ–Ω—Ç'
        elif ent_type == 'QUANTITY':
            ent_type = '–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ'
        elif ent_type == 'ORDINAL':
            ent_type = '–ü–æ—Ä—è–¥–∫–æ–≤–æ–µ —á–∏—Å–ª–æ'
        elif ent_type == 'CARDINAL':
            ent_type = '–ö–æ–ª–∏—á–µ—Å—Ç–≤–µ–Ω–Ω–æ–µ —á–∏—Å–ª–æ'
        else:
            ent_type = '–î—Ä—É–≥–æ–µ'

        # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –Ω–µ —è–≤–ª—è–µ—Ç—Å—è –ª–∏ —Å—É—â–Ω–æ—Å—Ç—å —á–∞—Å—Ç—å—é –¥—Ä—É–≥–æ–π —Å—É—â–Ω–æ—Å—Ç–∏
        is_part_of_other = False
        for other_ent in doc.ents:
            if other_ent != ent and ent.start >= other_ent.start and ent.end <= other_ent.end:
                is_part_of_other = True
                break

        if not is_part_of_other:
            entities.append((ent.text, ent_type))

    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –æ—Å–Ω–æ–≤–Ω—É—é —Ç–µ–º—É –¥–æ–∫—É–º–µ–Ω—Ç–∞
    noun_freq = Counter([token.lemma_ for token in doc if token.pos_ == 'NOUN'])
    top_nouns = [item[0] for item in noun_freq.most_common(3)]
    topic = ", ".join(top_nouns)

    # –û–ø—Ä–µ–¥–µ–ª—è–µ–º –æ—Å–Ω–æ–≤–Ω—ã–µ –¥–µ–π—Å—Ç–≤–∏—è, –æ–ø–∏—Å–∞–Ω–Ω—ã–µ –≤ –¥–æ–∫—É–º–µ–Ω—Ç–µ
    actions = [token.lemma_ for token in doc if token.pos_ == 'VERB']

    # –§–æ—Ä–º–∏—Ä—É–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç
    result = {
        "entities": entities,
        "topic": topic,
        "actions": actions
    }

    return result


def analyze_folder(folder_path):
    """
    –§—É–Ω–∫—Ü–∏—è –∞–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç –≤—Å–µ —Ñ–∞–π–ª—ã –≤ —É–∫–∞–∑–∞–Ω–Ω–æ–π –ø–∞–ø–∫–µ –∏ –≤–æ–∑–≤—Ä–∞—â–∞–µ—Ç —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã.
    """
    file_texts = preprocess_files(folder_path)
    results = {}
    for filename, text in file_texts.items():
        result = extract_document_meaning(text)
        results[filename] = result
    return results

# –£–∫–∞–∂–∏—Ç–µ –ø—É—Ç—å –∫ –ø–∞–ø–∫–µ Db
db_folder = "/content/Db"

all_results = analyze_folder(db_folder)

for filename, result in all_results.items():
    print(f"File: {filename}")
    print("Entities:")
    for entity, label in result["entities"]:
        print(f"- {entity} ({label})")
    print("Topic:", result["topic"])
    print("Actions:", ", ".join(result["actions"]))
    print()
```

```python
class VulnerabilityAssessor:
    def __init__(self):
        self.known_vulnerabilities = {
            "SQL Injection": ["select", "from", "where"],
            "Cross-Site Scripting (XSS)": ["<script>", "</script>"],
            "Insecure Configuration": ["debug=true", "admin_password="],
        }

    def assess_vulnerabilities(self, code):
        vulnerabilities = []
        for vuln_name, vuln_patterns in self.known_vulnerabilities.items():
            for pattern in vuln_patterns:
                if pattern in code.lower():
                    vulnerabilities.append(vuln_name)
                    break
        return vulnerabilities
```

```python
class DataClassifier:
    def __init__(self):
        self.classification_rules = {
            "Sensitive": ["ssn", "password", "credit_card"],
            "Confidential": ["trade_secret", "financial_data", "medical_records"],
            "Public": ["product_information", "press_release", "blog_post"],
        }

    def classify_data(self, data):
        classification_levels = []
        for level, keywords in self.classification_rules.items():
            for keyword in keywords:
                if keyword in data.lower():
                    classification_levels.append(level)
                    break
        return classification_levels
```

```python
# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –æ—Ü–µ–Ω—â–∏–∫–∞ —É—è–∑–≤–∏–º–æ—Å—Ç–µ–π
vulnerability_assessor = VulnerabilityAssessor()
code = "SELECT * FROM users WHERE id = 1; <script>alert('XSS attack');</script>"
vulnerabilities = vulnerability_assessor.assess_vulnerabilities(code)
print("Detected vulnerabilities:", vulnerabilities)

# –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä–∞ –¥–∞–Ω–Ω—ã—Ö
data_classifier = DataClassifier()
document_text = "This is a sensitive document containing your SSN: 123-45-6789."
classification_levels = data_classifier.classify_data(document_text)
print("Data classification:", classification_levels)
```
