```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
import os

def preprocess_files(directory, encryption_key):
    """
    Функция обрабатывает все файлы в указанной директории и во всех вложенных папках.
    Для PDF файлов извлекает текст, для PPTX файлов - извлекает текст из слайдов.
    Возвращает словарь, где ключи - зашифрованные имена файлов, а значения - зашифрованный извлеченный текст.
    """
    file_text = {}
    for root, dirs, files in os.walk(directory):
        for filename in files:
            if filename.endswith(".pdf"):
                pdf_path = os.path.join(root, filename)
                with open(pdf_path, 'rb') as pdf_file:
                    with PDF(pdf_file) as pdf:
                        text = "\n".join(page.extract_text() for page in pdf.pages)
                        encrypted_text = encrypt_text(text, encryption_key)
                        encrypted_filename = encrypt_filename(pdf_path, encryption_key)
                        file_text[encrypted_filename] = encrypted_text
            elif filename.endswith(".pptx"):
                pptx_path = os.path.join(root, filename)
                presentation = pptx.Presentation(pptx_path)
                text_slides = []
                for slide in presentation.slides:
                    text = "\n".join(shape.text for shape in slide.shapes if hasattr(shape, "text"))
                    text_slides.append(text)
                text = "\n\n".join(text_slides)
                encrypted_text = encrypt_text(text, encryption_key)
                encrypted_filename = encrypt_filename(pptx_path, encryption_key)
                file_text[encrypted_filename] = encrypted_text
    return file_text

def encrypt_text(text, key):
    cipher = AES.new(key, AES.MODE_GCM)
    ciphertext, tag = cipher.encrypt_and_digest(text.encode())
    return ciphertext, tag, cipher.nonce

def encrypt_filename(filename, key):
    cipher = AES.new(key, AES.MODE_GCM)
    ciphertext, tag = cipher.encrypt_and_digest(filename.encode())
    return f"{ciphertext.hex()}_{tag.hex()}_{cipher.nonce.hex()}"

# Пример использования
encryption_key = get_random_bytes(16)
db_directory = "/content/Db"
file_texts = preprocess_files(db_directory, encryption_key)
```

```python
from cryptography.hazmat.primitives.ciphers import Cipher, algorithms, modes
from cryptography.hazmat.backends import default_backend

def encrypt_data(data, key):
    """
    Шифрует данные, используя указанный ключ.
    """
    iv = get_random_bytes(16)
    cipher = Cipher(algorithms.AES(key), modes.GCM(iv), backend=default_backend())
    encryptor = cipher.encryptor()
    ciphertext = encryptor.update(data) + encryptor.finalize()
    tag = encryptor.tag
    return (ciphertext, tag, iv)

def decrypt_data(encrypted_data, key):
    """
    Расшифровывает данные, используя указанный ключ.
    """
    ciphertext, tag, iv = encrypted_data
    cipher = Cipher(algorithms.AES(key), modes.GCM(iv), backend=default_backend())
    decryptor = cipher.decryptor()
    decryptor.authenticate_additional_data(b"")
    return decryptor.update(ciphertext) + decryptor.finalize()
```

```python
import os
from datetime import datetime

def can_access_file(filename, file_size, file_mtime, key):
    """
    Проверяет, может ли пользователь с указанным ключом получить доступ к файлу.
    """
    # Проверка по имени файла
    if filename.startswith(key.decode()):
        return True

    # Проверка по размеру файла
    max_file_size = int(key[0:8])
    if file_size <= max_file_size:
        return True

    # Проверка по дате модификации файла
    max_mtime = int(key[8:16])
    file_mtime_dt = datetime.fromtimestamp(file_mtime)
    max_mtime_dt = datetime.fromtimestamp(max_mtime)
    if file_mtime_dt <= max_mtime_dt:
        return True

    return False
```

```python
def preprocess_files(directory, encryption_key, file_types=None):
    file_texts = {}
    for filename in os.listdir(directory):
        if file_types is None or any(filename.endswith(ft) for ft in file_types):
            file_path = os.path.join(directory, filename)
            file_size = os.path.getsize(file_path)
            file_mtime = os.path.getmtime(file_path)
            with open(file_path, 'rb') as file:
                file_data = file.read()
            encrypted_data = encrypt_data(file_data, encryption_key)
            file_texts[(filename, file_size, file_mtime)] = encrypted_data
    return file_texts

def manage_access(file_texts, access_permissions):
    accessible_files = {}
    for (filename, file_size, file_mtime), encrypted_data in file_texts.items():
        for user, key in access_permissions.items():
            if can_access_file(filename, file_size, file_mtime, key):
                decrypted_data = decrypt_data(encrypted_data, key)
                accessible_files[(filename, file_size, file_mtime)] = decrypted_data
                break
    return accessible_files
```

```python
import re

def extract_keywords(text):
    """
    Функция принимает текст из файла на русском языке и возвращает словарь ключевых слов с их частотой.
    """
    # Преобразуем текст в нижний регистр
    text = text.lower()

    # Удаляем знаки препинания
    text = re.sub(r'[^\w\s]', '', text)

    # Создаем список стоп-слов на русском языке
    stop_words = ['а', 'и', 'в', 'во', 'не', 'что', 'он', 'на', 'я', 'с', 'со', 'как', 'для', 'этот', 'быть', 'при', 'от', 'по', 'из', 'его', 'её', 'уже', 'какой', 'этого', 'себя', 'них', 'той', 'над', 'были', 'будет', 'ее', 'мы', 'тут', 'есть', 'нее', 'всех', 'будут', 'того']

    # Разбиваем текст на слова и фильтруем стоп-слова
    words = [word for word in re.findall(r'\w+', text) if word not in stop_words]

    # Формируем словарь ключевых слов с их частотой
    keywords = {}
    for word in words:
        if word in keywords:
            keywords[word] += 1
        else:
            keywords[word] = 1

    # Сортируем словарь по частоте в порядке убывания
    keywords = dict(sorted(keywords.items(), key=lambda item: item[1], reverse=True))

    return keywords
```

```python
import spacy
from collections import Counter
from spacy.tokens import Span

def extract_document_meaning(text):
    """
    Функция принимает текст документа и возвращает его основной смысл.
    """
    # Загружаем модель spaCy для русского языка
    nlp = spacy.load("ru_core_news_sm")

    # Обрабатываем текст с помощью spaCy
    doc = nlp(text)

    # Определяем ключевые сущности в тексте
    entities = []
    for ent in doc.ents:
        # Определяем тип сущности
        ent_type = ent.label_
        if ent_type == 'ORG':
            # Дополнительно проверяем, действительно ли это организация
            if any(token.text.lower() in ['компания', 'фонд', 'центр', 'институт', 'университет', 'завод', 'предприятие'] for token in ent):
                ent_type = 'Организация'
            else:
                ent_type = 'Другое'
        elif ent_type == 'LOC':
            ent_type = 'Местоположение'
        elif ent_type == 'PER':
            ent_type = 'Персона'
        elif ent_type == 'GPE':
            ent_type = 'Географический объект'
        elif ent_type == 'PRODUCT':
            ent_type = 'Продукт'
        elif ent_type == 'EVENT':
            ent_type = 'Событие'
        elif ent_type == 'WORK_OF_ART':
            ent_type = 'Произведение искусства'
        elif ent_type == 'LAW':
            ent_type = 'Закон'
        elif ent_type == 'MONEY':
            ent_type = 'Валюта/Денежная сумма'
        elif ent_type == 'DATE':
            ent_type = 'Дата'
        elif ent_type == 'TIME':
            ent_type = 'Время'
        elif ent_type == 'PERCENT':
            ent_type = 'Процент'
        elif ent_type == 'QUANTITY':
            ent_type = 'Количество'
        elif ent_type == 'ORDINAL':
            ent_type = 'Порядковое число'
        elif ent_type == 'CARDINAL':
            ent_type = 'Количественное число'
        else:
            ent_type = 'Другое'

        # Проверяем, не является ли сущность частью другой сущности
        is_part_of_other = False
        for other_ent in doc.ents:
            if other_ent != ent and ent.start >= other_ent.start and ent.end <= other_ent.end:
                is_part_of_other = True
                break

        if not is_part_of_other:
            entities.append((ent.text, ent_type))

    # Определяем основную тему документа
    noun_freq = Counter([token.lemma_ for token in doc if token.pos_ == 'NOUN'])
    top_nouns = [item[0] for item in noun_freq.most_common(3)]
    topic = ", ".join(top_nouns)

    # Определяем основные действия, описанные в документе
    actions = [token.lemma_ for token in doc if token.pos_ == 'VERB']

    # Формируем результат
    result = {
        "entities": entities,
        "topic": topic,
        "actions": actions
    }

    return result


def analyze_folder(folder_path):
    """
    Функция анализирует все файлы в указанной папке и возвращает результаты.
    """
    file_texts = preprocess_files(folder_path)
    results = {}
    for filename, text in file_texts.items():
        result = extract_document_meaning(text)
        results[filename] = result
    return results

# Укажите путь к папке Db
db_folder = "/content/Db"

all_results = analyze_folder(db_folder)

for filename, result in all_results.items():
    print(f"File: {filename}")
    print("Entities:")
    for entity, label in result["entities"]:
        print(f"- {entity} ({label})")
    print("Topic:", result["topic"])
    print("Actions:", ", ".join(result["actions"]))
    print()
```

```python
class VulnerabilityAssessor:
    def __init__(self):
        self.known_vulnerabilities = {
            "SQL Injection": ["select", "from", "where"],
            "Cross-Site Scripting (XSS)": ["<script>", "</script>"],
            "Insecure Configuration": ["debug=true", "admin_password="],
        }

    def assess_vulnerabilities(self, code):
        vulnerabilities = []
        for vuln_name, vuln_patterns in self.known_vulnerabilities.items():
            for pattern in vuln_patterns:
                if pattern in code.lower():
                    vulnerabilities.append(vuln_name)
                    break
        return vulnerabilities
```

```python
class DataClassifier:
    def __init__(self):
        self.classification_rules = {
            "Sensitive": ["ssn", "password", "credit_card"],
            "Confidential": ["trade_secret", "financial_data", "medical_records"],
            "Public": ["product_information", "press_release", "blog_post"],
        }

    def classify_data(self, data):
        classification_levels = []
        for level, keywords in self.classification_rules.items():
            for keyword in keywords:
                if keyword in data.lower():
                    classification_levels.append(level)
                    break
        return classification_levels
```

```python
import socket
import ssl

# Создаем SSL-контекст
context = ssl.create_default_context(ssl.Purpose.CLIENT_AUTH)
context.load_cert_chain(certfile="server.crt", keyfile="server.key")

# Создаем TCP-сокет и привязываем его к адресу
server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server_socket.bind(("localhost", 8443))
server_socket.listen(1)

print("SSL-сервер запущен и ожидает подключений...")

while True:
    # Принимаем входящее подключение
    client_socket, addr = server_socket.accept()

    # Обертываем сокет клиента SSL-контекстом
    ssl_client_socket = context.wrap_socket(client_socket, server_side=True)

    # Получаем данные от клиента
    data = ssl_client_socket.recv(1024)
    print(f"Получено от клиента: {data.decode()}")

    # Отправляем ответ клиенту
    ssl_client_socket.sendall(b"Привет, клиент!")

    # Закрываем соединение
    ssl_client_socket.shutdown(socket.SHUT_RDWR)
    ssl_client_socket.close()
```

```python
import socket
import ssl

# Создаем SSL-контекст
context = ssl.create_default_context()

# Подключаемся к SSL-серверу
with socket.socket(socket.AF_INET, socket.SOCK_STREAM, 0) as client_socket:
    ssl_client_socket = context.wrap_socket(client_socket, server_hostname="localhost")
    ssl_client_socket.connect(("localhost", 8443))

    # Отправляем данные серверу
    ssl_client_socket.sendall(b"Привет, сервер!")

    # Получаем ответ от сервера
    data = ssl_client_socket.recv(1024)
    print(f"Получено от сервера: {data.decode()}")
```
